```python
import cozmo
from matplotlib import pyplot as plt
from PIL import Image
import datetime
import cv2
import time
import numpy as np

'''
Go to object:
--> See bounding box 
--> Create custom object in the direction?
'''
def cozmo_program(robot: cozmo.robot.Robot):
    robot.set_lift_height(1.0)
    exposure_amount = 0.3 # Range: [0,1]
    gain_amount = 0.9 # Range: [0,1]
#     configure_camera(robot, exposure_amount, gain_amount)
    robot.camera.color_image_enabled = True
    robot.camera.image_stream_enabled = True
    
    i = 0
    while True: 
        latest_image = robot.world.latest_image
        if latest_image is not None:
            annotated = latest_image.annotate_image()

            plt.imshow(np.asarray(annotated))
            plt.pause(0.001) # imshow needs time to plot image
            print("image = %s" % latest_image)
            print('Do you want to save this image? (y for yes, s for skip):')

#             response = sys.stdin.readline()
            response = input()
            if response.strip() == 'y':
                fileName = 'cozmo_pics/'
                currentDT = datetime.datetime.now()
                i+=1
                fileName += 'image'+str(i)
                converted = annotated.convert()
                converted.save(fileName, "JPEG", resolution=10)
            else:
                continue
```

```python
# threading.Thread(target=saveImages).start()

cozmo.run_program(cozmo_program, use_viewer=False)
```
